```python
from selenium import webdriver
from bs4 import BeautifulSoup as BS
import time
from selenium.webdriver.common.keys import Keys
import re

driver = webdriver.Chrome()
driver.get("https://www.google.com.tw/?gfe_rd=cr&dcr=0&ei=0FXwWYnVOY7M8gfE5KXYBg")
driver.find_element_by_name('q').send_keys('台灣航空首頁',Keys.RETURN)

for p in range(13):
    soup = BS(driver.page_source,'lxml')
    a=0    
    for ele in soup.select('#rso a h3'):
        data_price=soup.select('.iUh30')[a].text
        a+=1
        m=re.search('https://(.*?)',data_price)   
        if m==None :
            data_price ='https:'+ data_price
        print (data_price)
        print (ele.text) 
        driver.find_element_by_link_text('下一頁').click()
    time.sleep(1)
driver.close()
```
